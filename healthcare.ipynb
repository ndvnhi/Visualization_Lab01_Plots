### ***Heatmaps***

```python
plt.figure(figsize = (10, 8))
sns.set_context('paper', font_scale=1.4)

hs_mx = hds_df.corr()
sns.heatmap(hs_mx, annot = True, cmap = 'Blues')
```

```python
hdh_bmi_pv = hds_df.pivot_table(index = ['heart_disease', 'hypertension'], columns = ['stroke'], values = 'bmi')
hdh_bmi_pv
```

```python
sns.heatmap(hdh_bmi_pv, cmap = 'Blues', linecolor = 'w', linewidths = 1)
```

```python
wr_age_pv = hds_df.pivot_table(index = ['work_type', 'Residence_type'], columns = ['stroke', 'hypertension'], values = 'age')
wr_age_pv
```

```python
sns.heatmap(wr_age_pv, cmap = 'Greens', linecolor = 'w', linewidths = 1)
```

### ***Facet Grid***

```python
hds_fg = sns.FacetGrid(hds_df, col = 'Residence_type', row = 'work_type', hue = 'stroke', height = 4, aspect = 1.3)
hds_fg.map(plt.scatter, 'bmi', 'avg_glucose_level', edgecolor = 'w')
```

```python
hds2_fg = sns.FacetGrid(hds_df, col = 'work_type', height = 3)
hds2_fg.map(plt.plot, 'bmi', 'avg_glucose_level', marker = '.')
```

### ***Regression Plot***

```python
plt.figure(figsize=(10, 8))
sns.set_context('paper', font_scale=1.4)
sns.lmplot(data = hds_df, x = 'bmi', y = 'avg_glucose_level', hue = 'stroke', markers = ['o', '^'],
           scatter_kws={'s': 30, 'linewidth': 0.5, 'edgecolor': 'w'})
```

```python
sns.lmplot(data = hds_df, x = 'bmi', y = 'avg_glucose_level', hue = 'stroke',
           col = 'hypertension', row = 'smoking_status')
```
